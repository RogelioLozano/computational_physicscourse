```python
# %load ../firstcell.py
%reload_ext autoreload
%autoreload 2

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='ticks')
plt.rcParams['figure.figsize'] = (20,10)
```

```python
def logistic(xn,lamb):
    return lamb*xn*(1-xn)
```

```python
l = np.linspace(1,4,1000)
N=1000
results = []

for lamb in l:
    initial_set = np.random.uniform(size=1000)
    for n in range(N):
        initial_set = logistic(initial_set,lamb)
    results.append(initial_set)
results = np.array(results,float)
```

```python
plt.plot(l,results,'o',markersize=.1);
sns.despine()
plt.xlabel('')
plt.ylabel('X')
```
