```python
import numpy as np
import matplotlib.pyplot as plt
from nested import nested_mult
from jupyterthemes import jtplot
jtplot.style()
```

```python
def newtdd(x,y):
    """Newton Divided Difference Interpolation Method.
    Computes coefficients of interpolating polynomial
    Input: x and y are arrays containing x and y coordinates of the n data points
    Output: Coefficients c of interpolating polynomial in nested form"""
    n=len(x)
    v = np.zeros((n,n))
    for j in range(n):
        v[j,0] = y[j]
    for i in range(1,n):
        for k in range(n-i):
            v[k,i] = (v[k+1,i-1]-v[k,i-1])/(x[k+i]-x[k]) 
    c = []
    for i in range(n):
        c.append(v[0,i])
    return np.array(c)
```

```python
x0=[0, 2, 3]
y0=[1, 2, 4]
coeff=newtdd(x0,y0)
```

```python
poly=nested_mult(coeff=coeff,b=[0,2])
```

```python
x=np.arange(-2,4,0.01)
plt.plot(x0,y0,'o',markersize=7)
plt.plot(x,poly(x))
```
