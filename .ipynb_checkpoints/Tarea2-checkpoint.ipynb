```python
def Legendre_calc(x,l,m):
    """Assumes x=cos(theta) float, l and m integers corresponding to associated legendre
    functions (l>0,m=-l,-l+1,...,l-1,l)"""
#     print("l=",l,"m=",m)
    if m<0:
#         print("Entrando m negativa")
        m=-m
        return (-1)**m * factorial(l-m)/factorial(l+m) * Legendre_calc(x,l,m) 
    if l == m:
#         print("Entrando l=m")
        Pll = ((-1)**l) * factorial2(2*l-1) * (1-x**2)**(l/2)
        return Pll
    elif l == m+1:
#         print("Entrando l=m+1")
        return x*(2*(l-1)+1)*Legendre_calc(x,l-1,m)
    else:
#         print("Entrando recursividad")
        L = (Legendre_calc(x,l-1,m)*x*(2*l-1) -(l+m-1)*Legendre_calc(x,l-2,m))/(l-m)
        return L
```

```python
def order_Legendre(l,m):
    def Legendre_curried(theta):
        return Legendre_calc(theta,l,m)
    return Legendre_curried
```

```python
P={}
for l in range(4):
    P[f"P{l}{0}"] = order_Legendre(l,0)
```

```python
fig,ax = plt.subplots()
x= np.cos(np.linspace(0,2*np.pi,200))
for poly in P.values():
    ax.plot(x,poly(x))
```

```python
def Spharmonics(theta,phi,l,m):
    coeff = np.sqrt( ((2*l+1)*factorial(l-m))/(factorial(l+m)*4*np.pi) )
    complex_spherical = coeff*order_Legendre(l,m)(np.cos(theta))*np.exp(1j*m*phi)
    return complex_spherical.real
```

```python
def raices_cuadratica(a,b,c):
    sol = lambda i: (-b+i*np.sqrt(b**2-4*a*c))/(2*a)
    return (sol(1),sol(-1))
```

La forma correcta no es solo multiplicar por el conjugado, lo que se debe hacer es considerar dos posibles casos, a saber, cuando b>0 o b<0:

```python
def nueva_cuadratica(a,b,c):
    sol1 = lambda i: (-b+i*np.sqrt(b**2-4*a*c))/(2*a)
    sol2 = lambda i: (2*c)/(-b+i*np.sqrt(b**2-4*a*c))
    if b > 0:
        return (sol1(-1),sol2(-1))
    if b<0:
        return (sol1(1),sol2(1))
```

Lo que sucede es una perdida de significancia al hacer diferencias en las soluciones de la cuadratica. Los errores de redondeo son inevitables cuando las direcciones de memoria de precisión finita se usan para representar número de precisión infinita en los reales. 

De acuerdo con el modelo de computo del estándar del IEEE 754 para aritmética en coma flotante, la representación de los números reales consiste de un conjunto de binario. Los numeros de punto flotante consisten de tres partes: el signo(+,-), la mantisa y el exponente. Estas tres partes se almacenan en una sola word.

--------------------
Una word es una pedazo de información de tamaño fijo que se maneja como unidad por el conjunto de instruciones i.e. el hardware del procesador. El número de bits en una word es una caraterística que define a los procesadores, en los actuales se tiene comunmente de 32 o 64 bits.

Las words se suelen usar para:

1. Números de punto fijo: integers
2. Números de punto flotante.
3. Direcciones (Addresses).
4. Registros de procesos.
5. Procesos de transferencia de memoria.

-----------------

Regresando al modelo, hay tres niveles de precisión para floats: single precision, double precision y long-double precision. El número de bits asignados para cada nivel son 32,64 y 80 respectivamente, estos se dividen en las partes mencionadas. Por ejemplo para double precision se tiene 1 bit para signo, 11 para el exponente y 52 para la mantisa.

La forma como se almacenan es $$\pm 1.bb...b x 2^p$$

Cada b es 0 o 1 y p es un numero binario de M bits representando el exponente.

Ejemplo: el 9 en binario es 1001 y seria almacenado como $+1.001x2^3$

La mayoria de compiladores basados en C usan double precision, el numero 1 con doble precision es: $$+1.0000000000000000000000000000000000000000000000000000x2^0$$

Con 52 bits de la mantisa. El siguiente numero flotante mayor a 1 es: $$+1.0000000000000000000000000000000000000000000000000001x2^0=1+2^{-52}$$

Definimos el epsilon de maquina como $\epsilon_{mach}$ como la distancia entre 1 y el menor numero de punto flotante mayor que 1. Para la IEEE double precision floating point standard es $$\epsilon_{mach} = 2^{-52}$$

El numero decimal 9.4 que en binario es 1001.011001100110..., se representa:
$$+1. 0010110011001100110011001100110011001100110011001100||110 . . . × 2^3$$

donde se puso doble barra despues de los primeros 52 bits de la mantisa. Aqui la explicacion de porque se tiene un error de redondeo para representar este tipo de numeros reales en las direcciones de memoria de precision finita de la computadora se puede empezar con la pregunta: Como ajustamos al numero 9.4 que en binario es infinito a un numero finito de bits?

Debemos truncar el numero de alguna manera, una manera es cortar los bits despues del 52, el protocolo es sencillo pero esta sesgado pues se mueven siempre los resultados hacia el cero.

El metodo alternativo es redondeo, se suma 1 si el binario despues del 52 es 1 y se corta o suma cero si es 0, con una excepcion: si los bits despues del bit 52 son 100000... que en decimal es 0.5, entonces redondeamos de acuerdo a cual de los dos posibles opciones hace que el bit final 52 sea 0.

Entonces el numero de punto flotante que representa 9.4 es $$+1. 0010110011001100110011001100110011001100110011001101 × 2^3$$

Llegamos a lo anterior descartando la cola infinita de $.1100x2^{-52}x2^3=.0110x2^{-51}x2^3=.4x2^{-48}$ de la derecha del numero y luego sumando $2^{-52}x2^3=2^{-49}$ en el paso de redondeo. Entonces:

$$floatRepres(9.4) = 9.4 + 2^{−49} − 0.4 × 2^{−48}= 9.4 + (1 − 0.8)2^{−49}= 9.4 + 0.2 × 2^{−49} $$

Es decir que una computadora usando el una representacion de float con precision doble y usando la regla de redondeo genera un error de $0.2 × 2^{−49}$ cuando almacena 9.4. Se llama a eso el error de redondeo.

Por lo tanto podemos ver de todo esto que al restar dos numero casi iguales, hay una perdida de las cifras significativas acumula un error grande en comparacion.

```python
from numpy import sign
def rootsearch(f,a,b,dx):
    """x1,x2 = rootsearch(f,a,b,dx).
    Searches the interval (a,b) in increments dx for
    the bounds (x1,x2) of the smallest root of f(x).
    Returns x1 = x2 = None if no roots were detected."""
    
    x1 = a; f1 = f(a)
    x2 = a + dx; f2 = f(x2)
    while sign(f1) == sign(f2):
        if x1>=b:
            return None,None
        x1 = x2; f1 = f2
        x2 = x1 + dx; f2 = f(x2)
    else:
        return x1,x2
    
def bisection(f,a,b,epsilon=0.5e-7):
    """Computes approximate solution of f(x)=0 using bisection method
    Input: function f, a,b, f(a)*f(b)<0 and tolerance epsilon
    Output approximate solution  interval (xc-epsilon,xc+epsilon)"""
    assert f(a)*f(b) < 0,"Please make sure that the interval contains a root"
    while (b-a)/2 > epsilon:
        c = (a+b)/2
        if f(c)==0:
            return c
        if f(a)*f(c) < 0:
            b=c
        else:
            a=c
    return a,b

def newton(f,df,x0,epsilon=0.5e-7):
    """Computes approximate solution of f(x)=0 using newton's method
    Input: function f, its derivative df, an initial guess x0 and tolerance epsilon
    Output approximate solution xc"""
    x1 = x0 - f(x0)/df(x0)
    while abs(x1-x0) > epsilon:
        x0=x1
        x1 = x0 - f(x0)/df(x0)
    return x1

def secante(f,x0,x1,epsilon=0.5e-7):
    """Computes approximate solution of f(x)=0 using secante method
    Input: function f, initial guesses x0 and x1 and tolerance epsilon
    Output approximate solution xc"""
    x2 = x1 - (f(x1)*(x1-x0))/(f(x1)-f(x0))
    while abs(x2-x1)>epsilon:
        x0 = x1
        x1 = x2
        x2 = x1 - (f(x1)*(x1-x0))/(f(x1)-f(x0))
    return x2

def ridder_root(f,a,b,tol=0.5e-7):
    
    assert f(a)*f(b) < 0, "The interval does not have a root"
    fa = f(a)
    fb = f(b)
    if fa == 0.0:
        return a
    if fb == 0.0:
        return b
    
    def processvalues(f,a,b):
        x3 = (a+b)/2
        fx3 = f(x3)
        dx = x3-a
        if (fa-fb)<0:
            dx = -dx
        x4 = x3 + dx*(fx3/np.sqrt(fx3**2 -fa*fb ))
        fx4 = f(x4)
        return x3,fx3,x4,fx4
    x3,fx3,x4,fx4 = processvalues(f,a,b)
    x_ant=x3
    
    while abs(x4-x_ant) > tol*max(abs(x4),1.0):
        if np.sign(fx3) == np.sign(fx4):
            if np.sign(fa)!= np.sign(fx4): b = x4; fb = fx4
            else: a = x4; fa = fx4
        else:
            a = x3; b = x4; fa = fx3; fb = fx4
        x_ant = x4
        x3,fx3,x4,fx4 = processvalues(f,a,b)
        
    return x4
```

```python
def potencial(phi,lamb,m):
    return -0.5*m**2*phi**2+lamb*phi**4/factorial(4)
```

```python
phi = np.linspace(-5,5)
fig = plt.subplots(figsize=(20,10))
plt.plot(phi,potencial(phi,1,1))
plt.axhline()
```

```python
def potencial_termico(T,phi,lamb,m):
    a4 = lamb/factorial(4) - 3*lamb/(512*np.pi**2) - lamb**2 * (np.log(2*m**2/(16*np.pi**2*T**2))-3/2+np.euler_gamma)/(256*np.pi**2)
    
    a2 = lamb * T**2 / 48 - m**2/2 +7*lamb*m**2/(128*np.pi**2) + lamb*m**2*(-3/2 + 2*np.euler_gamma)/(64*np.pi**2) - m**2 * lamb * (1+np.euler_gamma)/(32*np.pi**2) + T*lamb*np.sqrt(T**2 * lamb / 24 - m**2)/(16*np.pi)
    
    termex = T*( (-m**2+lamb*T**2/24)**(3/2) - (-m**2 + lamb*phi**2/2+lamb*T**2/24)**(3/2))/(12*np.pi)
    
    return a4*phi**4 + a2*phi**2 + termex 
```

```python
def curry_potential(T,lamb,m):
    def potential_curried(phi):
        return potencial_termico(T,phi,lamb,m)
    return potential_curried
```

```python
phi = np.linspace(0,15,1000)
T = 120.56
fig = plt.subplots(figsize=(20,10))
plt.plot(phi,potencial_termico(T,phi,1,np.sqrt(600)))
plt.axhline()
```

```python
points=1000
Tmax = 120.6

y=[]
temp = np.linspace(120.4,Tmax,points)

a,b = 6,10
for T in temp:
    try:
        therm_pot = curry_potential(T,1,np.sqrt(600))
        c = secante(therm_pot,a,b)
        y.append(c)
        a,b = c-.1,c+.1
    except AssertionError:
        break
y= np.array(y)
```

```python
fig = plt.subplots(figsize=(20,10))
plt.plot(temp[:len(y)],y)
plt.xlabel("Temperature",fontsize=20)
plt.ylabel("$\phi$",fontsize=20)
plt.axhline()
```

```python
index_critico=np.where(y==y.max())
Tcritica=temp[index_critico]
```
