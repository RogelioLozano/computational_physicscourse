```python
def binaria(x,k=3):
    if type(x) == int:
        if x < 0:
            x = -x
            print("La lista de output debe considerarse como un numero binario negativo")
        binario = []
        while x > 0:
            binario.append(x%2)
            x = x//2
        binario.reverse()
        return binario
    else:
        binario_entera = binaria(int(x),k)
        fraccion = x-int(x)
        binario_fraccion = []
        for i in range(k):
            fraccion = fraccion*2
            ent = int(fraccion)
            binario_fraccion.append(ent)
            fraccion = fraccion-ent
        return binario_entera + binario_fraccion
```

```python
def decimal(num_binario):
    d=0
    for j,n in enumerate(num_binario):
        j = len(num_binario)-(j+1)
        print("n:",n,"j:",j)
        d += n*(2**j)
    return d
```

Problema 4

#### En todos los incisos excepto 1 se usa el metodo de newton pues es rapido en convergencia, sencillo de programar, y al ser todas funciones analiticas en 1 dimension donde se pueden conocer sus raices y encontrar sus derivadas, usamos este metodo para todas las que se puede.

```python
#i)
poli = lambda x: 2*x**2+5*x-3
x= np.linspace(-4,2)
plt.plot(x,poli(x))
plt.axhline();
```

```python
# De la grafica como podemos ver que existen 2 raices entonces podemos separar el algoritmo
# en 2 espacios de busqueda distintos y obtener sus raices, usamos el metodo de newton por
# ser rapido en su convergencia

def newton(f, df, x, epsilon=1e-10):
    ff = f(x)
    res = [[x, ff, np.nan]]
    xn = x
    while True:
        xnp1 = xn - ff / df(xn)
        ff = f(xnp1)
        res.append([xnp1, ff, xnp1-xn])
        if abs(ff) < epsilon:
            break
        xn = xnp1
    return res

def Dpoli(x):
    return 4*x + 5

#Primera raiz
res = newton(poli,Dpoli,-4)
pd.DataFrame(res, columns=["$x_n$", "$f(x_n)$", "$x_n - x_{n-1}$"])
```

```python
#Segunda raiz
res = newton(poli,Dpoli,1)
pd.DataFrame(res, columns=["$x_n$", "$f(x_n)$", "$x_n - x_{n-1}$"])
```

```python
def Dquinta_func(x):
    if x<0:
        return 1/(3*-(-x)**(2.0/3))
    else:
        return 1/(3*x**(2.0/3))
    
def rootsearch(f,a,b,dx):
    x1 = a; f1 = f(a)
    x2 = a + dx; f2 = f(x2)
    while np.sign(f1) == np.sign(f2):
        if x1>=b: return None,None
        x1 = x2; f1 = f2
        x2 = x1 + dx; f2 = f(x2)
    else:
        return x1,x2
    
Dquinta_func = np.vectorize(Dquinta_func)

x1 = 0.0; x2 = 1.0
for i in range(5):
    dx = (x2 - x1)/10.0
    x1,x2 = rootsearch(quinta_func,x1,x2,dx)
x = (x1 + x2)/2.0
print('x =', '{:6.4f}'.format(x))
```
